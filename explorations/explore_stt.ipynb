```python
input_info = sd.query_devices(kind="input")

device_name = input_info["name"]
device_id = input_info["index"]
device_samplerate = int(input_info["default_samplerate"])
input_channels = int(input_info["max_input_channels"])

device_name, device_id, device_samplerate, input_channels
```

```python
from faster_whisper import WhisperModel
import sounddevice as sd
from scipy.io.wavfile import write
import os 
import sys 
import contextlib

# OUTPUT_DEVICE_ID = 5
# INPUT_DEVICE_ID = 1

whisper_model = WhisperModel("base", device="cpu", compute_type="int8")

duration = 5
with open(os.devnull, "w") as devnull:
    with contextlib.redirect_stderr(devnull):
        recording = sd.rec(int(duration*device_samplerate), samplerate=device_samplerate, channels=input_channels, device=device_id)
        print("Recording...")
        sd.wait()
        print("Recording finnished")
# time.sleep(duration+1)
write("recording.wav", device_samplerate, recording)
```

```python
def transcribe_audio(audio_path: str) -> str:
    segments, _ = whisper_model.transcribe(audio_path, language="sv")
    # Slå ihop alla segment till en sträng
    text = " ".join([segment.text for segment in segments])
    return text.strip()

text = transcribe_audio("recording.wav")
print(text)
```
